```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
# functions that returns dy/dt
def model(y, t, k):
    dydt = -k * y
    return dydt
```

```python
# initial condition
y0 = 5

# time points
t = np.linspace(0, 50)

# solve ODEs
k = [0.1, 0.2, 0.5]
y = np.zeros((len(k), t.size))
for i in range(len(k)):
    y[i, :] = odeint(model, y0, t, args=(k[i],)).T

# plot results
colors = ['r-', 'b--', 'g:']
for i in range(len(k)):
    plt.plot(t, y[i,:], colors[i], linewidth=2, label=f'k={k[i]}')
plt.xlabel('time')
plt.ylabel('y(t)')
plt.grid()
plt.legend()
plt.show()
```
