```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
# functions that returns dy/dt
def f1(y, t, k):
    dydt = -k * y
    return dydt

def f2(y, t, k):
    dydt = -y + k
    return dydt

def f3(y, t, k):
    # u steps from 0 to 2 at k=10 sec
    if t < k:
        u = 0
    else:
        u = 2
    dydt = (-y + u) / 5.0
    return dydt
```

```python
# First Problem
# initial condition
y0 = 5

# time points
t = np.linspace(0, 50)

# solve ODEs
k = [0.1, 0.2, 0.5] # list of multipliers
y = np.zeros((len(k), t.size))
for i in range(len(k)):
    y[i, :] = odeint(f1, y0, t, args=(k[i],)).T

# plot results
colors = ['r-', 'b--', 'g:']
for i in range(len(k)):
    plt.plot(t, y[i,:], colors[i], linewidth=2, label=f'k={k[i]}')
plt.xlabel('time')
plt.ylabel('y(t)')
plt.grid()
plt.legend()
plt.show()
```

```python
# Second Problem
# initial condition
y0 = 0

# time points
t = np.linspace(0, 50)

# solve ODEs
k = [1, 2, 5] # k: list of constants
y = np.zeros((len(k), t.size))
for i in range(len(k)):
    y[i, :] = odeint(f2, y0, t, args=(k[i],)).T

# plot results
colors = ['r-', 'b--', 'g:']
for i in range(len(k)):
    plt.plot(t, y[i,:], colors[i], linewidth=2, label=f'k={k[i]}')
plt.xlabel('time')
plt.ylabel('y(t)')
plt.grid()
plt.legend()
plt.show()
```

```python
# Third Problem
# initial condition
y0 = 1

# time points
t = np.linspace(0, 50)

# solve ODEs
k = [5.0, 10.0, 20.0] # k: list of times that input function has a rapid change
y = np.zeros((len(k), t.size))
for i in range(len(k)):
    y[i, :] = odeint(f3, y0, t, args=(k[i],)).T

# plot results
colors = ['r-', 'b--', 'g:']
for i in range(len(k)):
    plt.plot(t, y[i,:], colors[i], linewidth=2, label=f'k={k[i]}')
plt.xlabel('time')
plt.ylabel('y(t)')
plt.grid()
plt.legend()
plt.show()
```
